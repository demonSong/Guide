```python
# 统计去向
def totalLocationMap(featureName, fileName):
    df = pd.read_csv(fileName)
    datas = df[['userid', 'orderType', featureName]].values
    QLocalMap = {}
    allLocalMap = {}
    proMap = {}
    userSet = set()
    for data in datas:
        user = str(int(data[0]))
        orderType = str(data[1])
        add  = str(data[2])
        if add not in allLocalMap:
            QLocalMap[add] = 0
            allLocalMap[add] = 0
       
        if (orderType == '1'): QLocalMap[add] += 1
        allLocalMap[add] += 1
        userSet.add(user)
    for key in allLocalMap.keys():
        proMap[key] = QLocalMap[key] / allLocalMap[key]
    print(len(userSet))
    return QLocalMap, allLocalMap, proMap, userSet

# city 统计
QLocalMap, allLocalMap, proMap, userSet = totalLocationMap('city', './data/orderHistory_train.csv')
QLocalMap = sorted(QLocalMap.items(), key=lambda val:val[1],reverse=True)

proMap = sorted(proMap.items(), key=lambda val:val[1],reverse=True)
print(proMap)

def checkIfAllHistoryUserInTrain(userSetInHistory, fileName):
    train = pd.read_csv('./data/train.csv')
    datas = train['userid'].values
    cnt = 0
    for user in datas:
        user = str(int(user))
        if user not in userSetInHistory:
            cnt += 1
    if cnt == 0:
        print("userInHistory is all in train data!")
    else:
        print("sorry! userInHistory is not all in train data! The number is ", cnt)

checkIfAllHistoryUserInTrain(userSet, './data/train.csv')
```
